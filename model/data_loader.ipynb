```python
data_path = '../dataset/train/'
```

```python
from torchvision import transforms as T
```

```python
transform = T.Compose([
    T.Resize((256,256)),
    T.CenterCrop(224),
    T.ToTensor()
])
```

```python
from torchvision.datasets import ImageFolder
```

```python
train_dataset = ImageFolder(
    root=data_path,
    transform=transform
)
```

```python
import torch
```

```python
train_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=16,
    num_workers=0,
    shuffle=True
)
```

```python
len(train_loader)
```

```python
for idx, (inputs, targets) in enumerate(train_loader):
    for input_, target in zip(inputs, targets):
        if target == 4:
            img = T.ToPILImage()(input_)
            break
    break
```
